```python
def bar_plot(col , xlabel , ylabel , title):
    # Visualize Better and very good condition
    value_condition = df[col].value_counts()

    # Create the bar plot
    sns.barplot(x=value_condition.index, y=value_condition, palette="mako")

    # Add labels and title
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()
```

```python
def encoder(col):
    encoder = LabelEncoder()
    df[col] = encoder.fit_transform(df[col])
```

```python
input_data = df.drop(columns=['booking_complete'])
output_data = df['booking_complete']
```

```python
input_data
```

```python
x , x_test , y , y_test = train_test_split(input_data , output_data , test_size=0.2 , random_state=42)
x_train , x_val , y_train , y_val = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
random = HistGradientBoostingClassifier(max_depth= 500  , random_state=100)
random.fit(x_train, y_train)
```

```python
y_predict_train = random.predict(x_train)
y_predict_val =   random.predict(x_val)
acc_train = accuracy_score(y_train , y_predict_train)
acc_val = accuracy_score(y_val , y_predict_val)
```

```python
report_random = classification_report(y_train, y_predict_train, target_names = ['0', '1'])
print(report_random)
```
